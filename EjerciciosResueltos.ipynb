```python
import matplotlib.pyplot as plt
from PIL import Image

pic = Image.open('gato.jpg')
type(pic)
pic
```

```python
pic_arr = np.asarray(pic)
pic_arr.shape
```

```python
plt.imshow(pic_arr)
```

```python
pic_red = pic_arr.copy()
# Muestra el canal rojo con un mapa de color gris
plt.imshow(pic_red[:, :, 0], cmap='gray')
```

```python
pic_red[:, :, 1] = 0  # Canal verde a cero
pic_red[:, :, 2] = 0  # Canal azul a cero
plt.imshow(pic_red)
```

```python
import cv2
img = cv2.imread('gato.jpg')
fix_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(fix_img)
```

```python
fix_gray = cv2.imread('gato.jpg', cv2.IMREAD_GRAYSCALE)
plt.imshow(fix_gray, cmap='gray')
```

```python
new_img = cv2.resize(fix_img, (700, 400))
plt.imshow(new_img)
```

```python
w_ratio = 0.5
h_ratio = 0.5
new_img = cv2.resize(fix_img, (0, 0), fx=w_ratio, fy=h_ratio)
plt.imshow(new_img)
```

```python
cv2.imwrite('newcat.jpg', new_img)
```

```python
blank_img = np.zeros(shape=(512, 512, 3), dtype=np.int16)
```

```python
cv2.rectangle(blank_img, pt1=(384, 50), pt2=(500, 150), color=(0, 255, 0), thickness=10)
plt.imshow(blank_img)
```

```python
# Círculo con borde
cv2.circle(blank_img, center=(100, 100), radius=50, color=(0, 0, 255), thickness=5)
# Círculo relleno
cv2.circle(blank_img, center=(250, 250), radius=30, color=(255, 0, 0), thickness=-1)
plt.imshow(blank_img)
```

```python
cv2.line(blank_img, pt1=(0, 0), pt2=(512, 512), color=(0, 100, 200), thickness=5)
plt.imshow(blank_img)
```

```python
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(blank_img, text='Hola', org=(10, 500), fontFace=font, fontScale=4, color=(255, 255, 255), thickness=2, lineType=cv2.LINE_AA)
plt.imshow(blank_img)
```
